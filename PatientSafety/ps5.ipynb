```python
import torch
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline
pd.set_option("max_colwidth", 250)

torch.cuda.is_available()
```

```python
# Creating list to append tweet data to
tweets_list = []
max_tweets = 20000
start_time = '2022-01-01'
end_time = '2022-11-26'
query = "patient safety"
# Using TwitterSearchScraper to scrape data and append tweets to list

for i,tweet in enumerate(sntwitter.TwitterSearchScraper(
        f'{query} since:{start_time} until:{end_time}').get_items()):
    if i>max_tweets:
        break
    tweets_list.append([tweet.id, tweet.date, tweet.content, tweet.retweetCount, tweet.likeCount, tweet.lang, 
        tweet.user.followersCount, tweet.user.friendsCount, tweet.user.location, tweet.hashtags])

# Creating a dataframe from the tweets list above
tweet_2022 = pd.DataFrame(tweets_list, columns=['id', 'date', 'text', 'retweetCount', 'likeCount', 'lang',
                'followersCount', 'friendsCount', 'location', "hashtags"])
tweet_2022.to_csv("tweet_2022.csv", index=False)
```
